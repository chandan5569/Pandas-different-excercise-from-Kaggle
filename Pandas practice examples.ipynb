```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(1,5,10)
y=x**2
y
plt.plot(x,y)
plt.xlabel('<- x axis ->')                     #Functional Aproach
plt.ylabel('<- y axis ->')
plt.title('practice')
```

```python
#Subplot approach

plt.subplot(1,2,1)
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
f=plt.figure()
a=f.add_axes([0.1,0.1,1,1])              #Main axes we are creating Left,Bottom,Width,Height
b=f.add_axes([0.2,0.5,0.4,0.4])
a.plot(x,y)
b.plot(y,x)
a.set_title('outer')
b.set_title('inner')           #In oop approach we use set_title, set_xlabel and set_ylabel
```

```python
import pandas as pd
import numpy as np
l=np.linspace(1,2,36).reshape(6,6)
y=pd.DataFrame(l,index=[1,2,3,4,5,6],columns=['a','b','c','d','e','key'])
y
```

```python
pd.merge(x,y,on='key')             #MERGING CONCEPT
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],              #Merging on two index
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],                 #JOINING CONCEPT
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```
